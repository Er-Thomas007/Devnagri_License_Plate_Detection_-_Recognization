```python
import subprocess

# Install necessary libraries
subprocess.run(['pip', 'install', 'opencv-python-headless', 'matplotlib', 'ultralytics'])

# Import libraries
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import YOLO

# Load YOLOv8 model (assuming you've already trained it or using a pretrained model)
model = YOLO('/content/best.pt')  # Replace 'best.pt' with your trained model path if necessary

# Function to process an image
def process_image(image_path):
    # Perform inference on the image
    results = model(image_path)

    # Access the first result from the list
    result = results[0]

    # Convert the image to a format that can be displayed
    img_with_detections = result.plot()

    # Display the image with detections
    cv2_imshow(img_with_detections)
    plt.show()

# Function to process video
def process_video(input_video_path, output_video_path):
    # Open the video file
    cap = cv2.VideoCapture(input_video_path)

    # Get the video frame width, height, and FPS
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create a VideoWriter object to save the output video
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can change the codec as needed
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform inference on the current frame
        results = model(frame)

        # Access the first result from the list
        result = results[0]

        # Convert the image to a format that can be displayed
        img_with_detections = result.plot()

        # Write the frame with detections to the output video
        out.write(img_with_detections)

    # Release the video capture and writer objects
    cap.release()
    out.release()

    # Display the saved output video path
    print(f"Output video saved at: {output_video_path}")

# Path to the input image or video file
input_image_path = '/content/_dsc6075-2.jpg'  # Replace with the path to your image
input_video_path = '/content/2103099-uhd_3840_2160_30fps.mp4'  # Replace with the path to your input video
output_video_path = '/content/output_video.mp4'  # Replace with the desired output path for the processed video

# Choose which type of processing to run
process_type = 'video'  # Set to 'video' for processing video, 'image' for processing image

if process_type == 'image':
    process_image(input_image_path)
elif process_type == 'video':
    process_video(input_video_path, output_video_path)

# If in a notebook environment, display the output video file
from IPython.display import HTML
from base64 import b64encode

def show_video(video_path, video_width):
    video_file = open(video_path, "rb").read()
    video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
    return HTML(f"""<video width="{video_width}" controls>
                        <source src="{video_url}" type="video/mp4">
                    </video>""")

# Display the output video (if processing a video)
if process_type == 'video':
    display(show_video(output_video_path, 640))  # Adjust width as needed
```
